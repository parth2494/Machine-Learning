```python
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(y='day', x='total_bill', data=tips, palette='rainbow', orient='h')
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow', hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, palette='rainbow')
sns.violinplot(x='day', y='total_bill', data=tips, palette='Set1')
```
